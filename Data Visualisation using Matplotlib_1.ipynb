## Bar plot
A bar plot or bar chart is a graph that represents the category of 
data with rectangular bars with lengths and heights that is proportional
to the values which they represent.

```python
# creating the dataset
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='k',
        width = 0.4)
 
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
## Object Oriented plotting

## It refers to user defined plotting where we plot anatomy of plots.
```

```python
import math
x=np.arange(0,math.pi*2,0.05)
```

```python
x
```

```python
y=np.sin(x)
```

```python
y
```

```python
## creating blank or empt canvas
fig=plt.figure()

##Addging axes
ax=fig.add_axes([0.5,0.5,0.5,0.5])

## labeling the axes
ax.set_title('Sine Wave')
ax.set_xlabel('Angle')
ax.set_ylabel('Sine')
ax.plot(x,y)
plt.show()
```

```python
y=np.cos(x)
```

```python
## creating blank or empt canvas
fig=plt.figure()

##Addging axes
ax=fig.add_axes([1,1,1,1])

## labeling the axes
ax.set_title('Cos Wave')
ax.set_xlabel('Angle')
ax.set_ylabel('Cos')
ax.plot(x,y)
plt.show()
```

```python
##Common layouts plots

fig,a=plt.subplots(2,2)
x=np.arange(1,6)
a[0][0].plot(x,x*x)
a[0][0].set_title('First Plot')
a[0][1].hist(x)
a[0][1].set_title('Histogram')
a[1][0].plot(x,np.sqrt(x))
a[1][0].set_title('Square Root')
a[1][1].plot(x,np.log(x))
a[1][1].set_title('Log')
plt.show()
```

```python
x=np.arange(0,math.pi*2)
y=np.tan(x)
## creating blank or empt canvas
fig=plt.figure()

##Addging axes
ax=fig.add_axes([0,0,1,1])

## labeling the axes
ax.set_title('Tan Wave')
ax.set_xlabel('Angle')
ax.set_ylabel('Tan')
ax.plot(x,y)
plt.show()
```
